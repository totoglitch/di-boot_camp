```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Data Preparation
np.random.seed(42)
temperatures = np.random.uniform(-5, 35, size=(10, 12))
cities = [f"City {i+1}" for i in range(10)]
months = [f"Month {i+1}" for i in range(12)]
df = pd.DataFrame(temperatures, index=cities, columns=months)

# 2. Data Analysis
annual_avg_temp = df.mean(axis=1)
highest_temp_city = annual_avg_temp.idxmax()
lowest_temp_city = annual_avg_temp.idxmin()

# 3. Data Visualization
plt.figure(figsize=(10, 6))
for city in cities:
    plt.plot(months, df.loc[city], label=city)

plt.xlabel("Month")
plt.ylabel("Average Temperature (°C)")
plt.title("Monthly Average Temperatures for 10 Cities")
plt.legend()
plt.grid(True)
plt.show()

# Summary Report
print(f"The city with the highest average temperature is: {highest_temp_city} ({annual_avg_temp[highest_temp_city]:.2f} °C)")
print(f"The city with the lowest average temperature is: {lowest_temp_city} ({annual_avg_temp[lowest_temp_city]:.2f} °C)")
print("\nFurther analysis could include finding trends in the temperature variation within each city across months, comparing temperature differences between cities, or identifying any outliers in the dataset. The visualisation provides a quick view of the temperature trends in each city across the year.")

```
