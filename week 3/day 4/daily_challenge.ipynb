```python
import pandas as pd

try:
    df = pd.read_csv('2019.csv')
    display(df.head())
except FileNotFoundError:
    print("Error: '2019.csv' not found. Please ensure the file exists in the current directory.")
    df = None  # Set df to None to indicate failure
except pd.errors.EmptyDataError:
    print("Error: '2019.csv' is empty.")
    df = None
except pd.errors.ParserError:
    print("Error: Unable to parse '2019.csv'. Please check the file format.")
    df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
```

```python
print("Shape of the DataFrame:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nDescriptive Statistics:\n", df.describe())
print("\nMissing Values per column:\n", df.isnull().sum())
print("\nUnique values in 'Country or region' column:\n", df['Country or region'].unique())
print("\nFrequency of unique values in 'Country or region' column:\n", df['Country or region'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df['region']= df['Country or region']
plt.scatter(df['Social support'], df['Score'])
plt.xlabel('Social support')
plt.ylabel('Score')
plt.title('Social support vs. Score')
plt.show()
```

```python
plt.scatter(df['GDP per capita'], df['Healthy life expectancy'])
plt.xlabel('GDP per capita')
plt.ylabel('Healthy life expectancy')
plt.title('GDP per capita vs. Healthy life expectancy')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Subplot 1: Bar plot for 'GDP per Capita'
axes[0].bar(df['region'], df['GDP per capita'])
axes[0].set_xlabel('Region')
axes[0].set_ylabel('GDP per Capita')
axes[0].set_title('GDP per Capita by Region')
axes[0].tick_params(axis='x', rotation=45)

# Subplot 2: Line plot for 'Healthy Life Expectancy'
axes[1].plot(df['region'], df['Healthy life expectancy'])
axes[1].set_xlabel('Region')
axes[1].set_ylabel('Healthy Life Expectancy')
axes[1].set_title('Healthy Life Expectancy by Region')
axes[1].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```
