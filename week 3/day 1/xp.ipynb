```python
Data analysis is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, informing conclusions,
and supporting decision-making.  It involves a systematic approach to understanding data, identifying patterns,
and extracting meaningful insights that can be used to solve problems or improve processes.
This can range from simple descriptive statistics to complex predictive modeling using sophisticated algorithms.
In today's data-driven world, data analysis is paramount.  The sheer volume of data generated daily across various sectors makes it
impossible to discern valuable insights without systematic analysis.  Businesses, governments, and researchers alike rely on data analysis to make informed decisions,
improve efficiency, identify opportunities, and mitigate risks.  Essentially, data analysis provides the lens through which we can understand our increasingly complex world,
enabling us to act with knowledge rather than speculation.
Three key areas where data analysis is applied today include:

1. **Business and Marketing:** Data analysis plays a vital role in optimizing marketing campaigns, understanding customer behavior,
and improving sales strategies. Businesses analyze customer demographics, purchase history, website traffic, and social media engagement to create targeted advertising campaigns,
personalize customer experiences, and predict future trends.  For example, e-commerce platforms leverage data analysis to recommend products, optimize pricing,
and personalize website layouts based on individual browsing habits.  This leads to increased customer satisfaction, conversion rates, and overall revenue.
2. **Healthcare:** Data analysis is revolutionizing healthcare through improved diagnostics, personalized treatments, and disease prediction.
Hospitals and research institutions utilize data analysis to analyze patient medical records, genetic information, and clinical trial data to identify disease patterns,
develop new treatments, and predict patient outcomes.  For instance,
analyzing large datasets of patient information can help identify individuals at high risk of developing certain diseases, enabling proactive interventions and early diagnosis.
Additionally, data analysis supports the development of drug discovery and personalized medicine, leading to more effective and tailored treatments.
3. **Finance:**  In the financial sector, data analysis is critical for risk management, fraud detection, and investment strategies.
Financial institutions analyze market trends, transaction patterns, and credit scores to assess risks, identify fraudulent activities,
and make investment decisions.  Algorithmic trading models, powered by data analysis, automate trading processes based on real-time market data,
enabling quicker and potentially more profitable trades. Moreover, credit scoring models utilize data analysis to assess an individual's creditworthiness,
allowing lenders to make more informed decisions about loan approvals and interest rates.

In conclusion, data analysis has become an indispensable tool across various sectors.
Its importance lies in the ability to extract meaningful insights from complex datasets,
facilitating better decision-making, improved processes, and a deeper understanding of our world.
As data generation continues to accelerate, the role of data analysis will only become more crucial in shaping our future.
```

Exercise 3: Identifying Data Types


Objective: Learn to identify different data types.
Task:
For the datasets from the previous exercise, categorize each column of it as either quantitative or qualitative and explain your reasoning.



```python
def identify_data_types(df, df_name):
    """
    Categorizes columns of a DataFrame as quantitative or qualitative.
    """

    print(f"\nData Type Identification for {df_name}:")
    for col in df.columns:
        dtype = df[col].dtype
        if pd.api.types.is_numeric_dtype(df[col]):
            data_type = "Quantitative"
        elif pd.api.types.is_datetime64_any_dtype(df[col]):
          data_type = "Quantitative"
        elif pd.api.types.is_categorical_dtype(df[col]):
            data_type = "Qualitative"
        elif dtype == 'object':
            data_type = "Qualitative"
        else:
            data_type = "Unknown"

        print(f"- Column '{col}': {data_type}")
identify_data_types(sleep_df, "Sleep Health Dataset")
identify_data_types(mental_health_df, "Mental Health Dataset")
identify_data_types(credit_df, "Credit Card Approval Dataset")
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
mean_sepal_length = iris_df['SepalLengthCm'].mean()
median_sepal_length = iris_df['SepalLengthCm'].median()
mode_sepal_length = iris_df['SepalLengthCm'].mode()[0]

print(f"Mean Sepal Length: {mean_sepal_length}")
print(f"Median Sepal Length: {median_sepal_length}")
print(f"Mode Sepal Length: {mode_sepal_length}")

plt.figure(figsize=(8, 6))
plt.hist(iris_df['SepalLengthCm'], bins=10, edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.show()
plt.figure(figsize=(8, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=iris_df)
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length Distribution by Species')
plt.show()
plt.plot(iris_df['SepalLengthCm'], iris_df['SepalWidthCm'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs. Sepal Width')
plt.show()
```
